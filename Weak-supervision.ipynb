```python
import pandas as pd
from rapidfuzz import fuzz
```

```python
def lf_skill_overlap(query_skills, job_skills):
    overlap = len(set(query_skills) & set(job_skills))
    
    if overlap >= 4:
        return 2   
    elif overlap >= 2:
        return 1   
    else:
        return 0   
```

```python
def lf_esco_match(query_esco, job_esco):
    if query_esco == job_esco:
        return 2
    else:
        return 0
```

```python
def lf_embedding_similarity(sim):
    if sim > 0.8:
        return 2
    elif sim > 0.65:
        return 1
    else:
        return 0
```

```python
def lf_title_similarity(query, job_title):
    score = fuzz.partial_ratio(query.lower(), job_title.lower())
    
    if score > 85:
        return 2
    elif score > 65:
        return 1
    else:
        return 0
```

```python
def combine_weighted(lf_outputs):
    skill, esco, emb, title = lf_outputs
    
    score = (
        0.4 * skill +
        0.3 * esco +
        0.2 * emb +
        0.1 * title
    )
    
    if score >= 1.5:
        return 2
    elif score >= 0.75:
        return 1
    else:
        return 0
```
